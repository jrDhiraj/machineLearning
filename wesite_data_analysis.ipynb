```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/data-export (1).csv')
```

```python
df.head()
```

```python
df.columns = df.loc[0]
```

```python
df.head()
```

```python
df.head()
```

```python
df = df.drop(index=0).reset_index(drop=True)
df.columns = ['channel_Groups', 'Date_hours', 'users', 'sessions', 'Engaged sessions','Average engagement time per session','Engaged sessions per user',	'Events per session',	'Engagement rate',	'Event count']
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date_hours'] = pd.to_datetime(df['Date_hours'], format='%Y%m%d%H', errors='coerce')
```

```python
df.head()
```

```python
numeric_columns = df.columns.drop(['channel_Groups','Date_hours'])

df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

df['Hours'] = df['Date_hours'].dt.hour
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(10,6))

df.groupby('Date_hours')[['sessions', 'users']].sum().plot(ax=plt.gca())
plt.title('Total Sessions and Users Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.barplot(data = df , x = 'channel_Groups', y = 'users')
plt.title('users by Channel Group')
plt.xlabel('Channel Group')
plt.xticks(rotation = 45)
plt.ylabel('users')
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.barplot(data = df, x='channel_Groups', y='Average engagement time per session', color='red')
plt.title('Average engagement time per session by Channel Group')
plt.xlabel('Channel Group')
plt.xticks(rotation=45)
plt.ylabel('Average engagement time per session')
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(data=df , x = 'channel_Groups', y = 'Engagement rate')
plt.title('Engagement rate by channel Groups')
plt.xlabel('channel groups')
plt.ylabel('Engagement rate')
plt.xticks(rotation=45)
plt.show()
```
